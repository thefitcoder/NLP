```
import nltk
nltk.download('all')
```

```
from nltk.corpus import brown
brown.categories()
```

```
brown.words(categories='adventure')[:10]
```

```
len(brown.words())
```

```
from nltk.corpus import inaugural
inaugural.fileids()
```

```
inaugural.words(fileids='2013-Obama.txt')
```

```
speech = ' '.join(inaugural.words(fileids='2013-Obama.txt'))
print(speech)
```

```
#Frequency Distribution
from nltk.probability import FreqDist
fd = FreqDist(speech.split(' '))
fd
```

```
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word), word) for word in speech.split()) 
cfd[5]
```
